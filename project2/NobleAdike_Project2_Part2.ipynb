### Data Quality Observations

Looking at the missing values, most columns are in good shape. The skill stat columns (Crossing, Dribbling, etc.) along with Height, Weight, Preferred Foot and others all have the same 48 rows missing. These are incomplete player profiles that we will need to drop entirely. Club has 241 missing values which just means those players are free agents, so we fill those with "Free Agent". Loaned From has the most missing values at 16,943 (93%) but this is expected as it only applies to players currently on loan, so the rest can be filled with "Not on Loan". Position and Jersey Number have 60 missing values each  and we can fill with "Unknown" and the median respectively. Joined, Contract Valid Until, and Release Clause might need to be left missing or dropped entirely since these cannot be reliably guessed or imputed. No duplicate rows were found in the dataset.


## Strategy to handle missing data for dataset
* 48 rows missing all skill stats + Height/Weight/Preferred Foot so dropping these rows entirely (incomplete profiles, not salvageable)
* Club column (241) — fill with "Free Agent"
* Loaned From (16,943) — fill with "Not on Loan" (it's structural, not missing data)
* Joined / Contract Valid Until / Release Clause — leave as NaN (can't reliably impute contract dates/clauses)

```python
#Make a copy of the original dataset to work with for cleaning
df_cleaned = df.copy()
```


Dropping the 48 rows where skill stats, Height, Weight, and Preferred Foot are all missing

```python
rows_before = len(df_cleaned)

df_cleaned = df_cleaned.dropna(subset=['Crossing'])

rows_after = len(df_cleaned)
print(f"Dropped: {rows_before - rows_after} rows")
print(f"Remaining rows: {rows_after}")
```

241 players have no club (free agents) so we filling  with "Free Agent" to keep their rows.

```python
df_cleaned['Club'] = df_cleaned['Club'].fillna('Free Agent')
print("Club missing after fill:", df_cleaned['Club'].isna().sum())
```

Only ~7% of players are on loan, so `Loaned From` is structurally empty for the rest so we can fill with "Not on Loan".

```python
df_cleaned['Loaned From'] = df_cleaned['Loaned From'].fillna('Not on Loan')
print("Loaned From missing after fill:", df_cleaned['Loaned From'].isna().sum())
```

```python
print('Position missing:', df_cleaned['Position'].isna().sum())
print('Jersey Number missing:', df_cleaned['Jersey Number'].isna().sum())
```


After the drop, 12 players still lack these. Fill `Position` with `"Unknown"` and `Jersey Number` with the median.

```python
df_cleaned['Position'] = df_cleaned['Position'].fillna('Unknown')
df_cleaned['Jersey Number'] = df_cleaned['Jersey Number'].fillna(df_cleaned['Jersey Number'].median())

print('Position missing:', df_cleaned['Position'].isna().sum())
print('Jersey Number missing:', df_cleaned['Jersey Number'].isna().sum())
```

### Converting currency columns to numeric

Stripping currency strinhgs and convert suffixes to float values in euros.

```python
def parse_currency(val):
    if pd.isna(val):
        return np.nan
    val = str(val).replace('€', '').strip()
    if 'M' in val:
        return float(val.replace('M', '')) * 1_000_000
    elif 'K' in val:
        return float(val.replace('K', '')) * 1_000
    else:
        return float(val)

df_cleaned['Value'] = df_cleaned['Value'].apply(parse_currency)
df_cleaned['Wage'] = df_cleaned['Wage'].apply(parse_currency)
df_cleaned['Release Clause'] = df_cleaned['Release Clause'].apply(parse_currency)

print(df_cleaned[['Value', 'Wage', 'Release Clause']].dtypes)
df_cleaned[['Value', 'Wage', 'Release Clause']].head()
```

#### Converting `Height` and `Weight` to numeric



```python
def parse_height(h):
    if pd.isna(h):
        return np.nan
    parts = str(h).replace('"', '').split("'")
    feet = int(parts[0])
    inches = int(parts[1]) if len(parts) > 1 and parts[1] else 0
    return round((feet * 12 + inches) * 2.54, 2)

def parse_weight(w):
    if pd.isna(w):
        return np.nan
    return float(str(w).replace('lbs', ''))

df_cleaned['Height_cm'] = df_cleaned['Height'].apply(parse_height)
df_cleaned['Weight_lbs'] = df_cleaned['Weight'].apply(parse_weight)

print(df_cleaned[['Height_cm', 'Weight_lbs']].describe())
```

### Converting date columns to datetime

```python
df_cleaned['Joined'] = pd.to_datetime(df_cleaned['Joined'].astype(str))
df_cleaned['Contract Valid Until'] = pd.to_datetime(df_cleaned['Contract Valid Until'].astype(str), errors='coerce')

print(df_cleaned[['Joined', 'Contract Valid Until']].dtypes)
```

#### Exporting the cleaned values and checking if there are more missing values (release clause, joined, contract valid are exceptions)

```python
print("=== Remaining missing values ===")
remaining = df_cleaned.isna().sum()
print(remaining[remaining > 0])
print(f"\nFinal shape: {df_cleaned.shape}")

df_cleaned.to_csv("fifa_eda_stats_cleaned.csv", index=False)
print("\nCleaned dataset saved as fifa_eda_stats_cleaned.csv")
```

### 6. Add a New Column: BMI

BMI (Body Mass Index) = weight in kg / (height in metres) squared. It gives a standardised measure of a player's body composition relative to their size, useful for comparing physical profiles across positions.

```python
# Weight_lbs → kg, Height_cm → metres
df_cleaned['Weight_kg'] = df_cleaned['Weight_lbs'] * 0.453592
df_cleaned['BMI'] = df_cleaned['Weight_kg'] / ((df_cleaned['Height_cm'] / 100) ** 2)

df_cleaned[['Name', 'Height_cm', 'Weight_lbs', 'Weight_kg', 'BMI']].head()
```

## 11. Visualizations

### Section I: Distribution Visualizations
### Which countries produce the most elite players?

#### Visualization: Count Plot — Countries ranked by how elite players produced (top 10 only)

This count plot for a categorical variable (Nationality) shows which countries produce the most elite players (players with an overall rating greater than 80)`.groupby()` is used to aggregate counts per nationality.

```python
top_player_counts = (
    df_cleaned[df_cleaned['Overall'] >= 80]
    .groupby('Nationality', as_index=False)
    .agg(Player_Count=('Overall', 'count'))
    .sort_values('Player_Count', ascending=False)
    .head(10)
)

plt.figure(figsize=(11, 6))
ax = sns.barplot(data=top_player_counts, x='Nationality', y='Player_Count', order=top_player_counts['Nationality'], palette='Blues_d')

for i, v in enumerate(top_player_counts['Player_Count']):
    ax.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Top 10 Nationalities with the most elite players (Overall ≥ 80)')
plt.xlabel('Nationality')
plt.ylabel('Number of Elite Players (Overall ≥ 80)')
plt.tight_layout()
#plt.savefig('top_countries_elite_players.png')
plt.show()
```

The diagram shows that Brazil, Spain, and France produce the highest number of top-rated players (Overall ≥ 80), reflecting strong footballing cultures and elite youth academies. England, despite having the largest total player count in the dataset, does not appear in the top 3 here which would confirm that quantity of players does not translate to proportionally high elite talent. 

### At what age do players tend to reach their peak overall rating?

#### GroupBy Visualization: Average Overall Rating by Age


```python
age_overall = (df_cleaned[df_cleaned['Age'] <= 40].groupby('Age')['Overall'].mean().reset_index())
age_overall.columns = ['Age', 'Avg_Overall']

plt.figure(figsize=(12, 6))
sns.lineplot(data=age_overall, x='Age', y='Avg_Overall', marker='o', color='blue', )

peak = age_overall.loc[age_overall['Avg_Overall'].idxmax()]
plt.axvline(peak['Age'], color='red', linestyle='--', label=f"Peak Age: {int(peak['Age'])}")
plt.legend()

plt.title('Average Overall Rating by Age')
plt.xlabel('Age')
plt.ylabel('Average Overall Rating')
plt.xticks(age_overall['Age'], rotation=45)
plt.tight_layout()
plt.show()
```

The average Overall rating rises steadily from the late teens and peaks around age 31 before dropping off sharply. The upward trend makes sense since younger players are still developing and tend to be rated lower. The peak at 31 is likely a result of survivorship bias, where only the very best players are still professionally active at that age, which inflates the group average. Players younger than 20 average below 65 while those in their late 20s and early 30s cluster around 70+. The sharp drop after 35 reflects just how rare it is for elite players to still be competing at the top level past that point.

### At what age do players tend which their highest market values?
#### Bar Chart — Average Market Value by Age


```python
age_value = df_cleaned[df_cleaned['Value'] > 0].groupby('Age')['Value'].mean().reset_index()
age_value['Value_M'] = age_value['Value'] / 1_000_000

plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=age_value, x='Age', y='Value_M', marker='o', color='darkorange')

peak = age_value.loc[age_value['Value_M'].idxmax()]
plt.axvline(peak['Age'], color='red', linestyle='--', label=f"Peak Age: {int(peak['Age'])}")
plt.legend()

plt.title('Average Market Value by Age (€ Millions)')
plt.xlabel('Age')
plt.ylabel('Average Market Value (€ Millions)')
plt.xticks(age_value['Age'], rotation=45)
plt.show()
```

Explanation of Visualization: Market value rises steadily through a player's early and mid-20s, peaking around age 27 when clubs are willing to pay the most for players in their prime. There is a small dip after 27 before a slight recovery at 31, which is likely explained by the fact that this is a 2018 dataset where several of the world's top players like Messi, Suarez and Ronaldo were around that age, pulling the average back up. After 31 the decline becomes steep and consistent so usually value drops as players approach their mid-30s and clubs become reluctant to invest heavily in aging players.

## Section II Visualization 

### Which skills tend to grow together in players?
#### Heatmap — Correlation of Core Technical Skills

This heatmap shows correlations among 10 technical skill attributes and reveal which skills tend to co-develop in the same player.

```python
technical_skills = [
    'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
    'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl'
]
corr_matrix = df_cleaned[technical_skills].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)

plt.title('Correlation Heatmap — Core Technical Skills')
plt.savefig('technical_skills_correlation.png')
plt.show()
```

Most technical skills show strong positive correlations, which makes sense since players who are good at one skill tend to be good at others too. The strongest pairs are ShortPassing and BallControl (r = 0.94), which are almost inseparable since both are core parts of possession-based technique, followed by Dribbling and BallControl (r = 0.90) since dribbling really depends on first-touch and ball retention. HeadingAccuracy stands out as the weakest correlator across the board (often r < 0.5), probably because aerial ability is more physically determined and less tied to the kind of technical development that drives the other skills. There are no strong negative correlations so overall these skills seem to broadly co-develop rather than trade off against each other.
